```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Carregar o dataset Wine
wine = load_wine()
X = wine.data
y = wine.target
```

```python
# Dividir entre treino e teste (70/30)
# Usamos stratify=y para garantir que as 3 classes de vinho
# estejam bem distribuídas no treino e no teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
# Criar o modelo Naive Bayes
modelo_nb = GaussianNB()
```

```python
# Treinar o modelo
modelo_nb.fit(X_train, y_train)
```

```python
# Fazer previsões
y_pred = modelo_nb.predict(X_test)
```

```python
# Avaliar o modelo
print("=== RESULTADOS - NAIVE BAYES NA BASE WINE ===\n")
print(f"Acurácia: {accuracy_score(y_test, y_pred):.4f}")

print("\nRelatório de Classificação:\n",
      classification_report(y_test, y_pred, target_names=wine.target_names))
```

```python
# Gráfico Matriz de Confusão
plt.figure(figsize=(6, 4))
cm = confusion_matrix(y_test, y_pred)
```

```python
# Criar matriz
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',
            xticklabels=wine.target_names,
            yticklabels=wine.target_names)

plt.title('Matriz de Confusão - Wine Dataset (Naive Bayes)')
plt.ylabel('Tipo de Vinho Real')
plt.xlabel('Tipo de Vinho Previsto')
plt.tight_layout()
plt.show()
```
